```python
import urllib.request
from bs4 import BeautifulSoup
base="https://search.kyobobook.co.kr/search?keyword=%ED%8C%8C%EC%9D%B4%EC%8D%AC&gbCode=TOT&target=total&page="
book_list=[]  
```

```python
import re
def save_image(src, title):
    pattern = r'[가-힣a-zA-Z0-9]+'  # 한글, 영문 대소문자, 숫자 포함
    result = re.findall(pattern, title)
    result=' '.join(result)
    save_path=f'images/{result}.jpg'
    urllib.request.urlretrieve(src,save_path)
    print(title,"이미지 저장")
```

```python
import urllib.request
from bs4 import BeautifulSoup
base="https://www.yes24.com/Product/Search?domain=BOOK&query=python&page="
book_list=[]

for i in range(3):
    url=base+str(i+1)
    html = urllib.request.urlopen(url)
    soup = BeautifulSoup(html, "html.parser")
    item_list=soup.select("#yesSchList > li")
    #print(item_list)
    k=0
    for item in item_list:
        title=item.select_one('a.gd_name')
        title_txt=title.text
        print(title_txt)
        
        author=item.select_one('span.authPub.info_auth > a')
        author_txt=author.text
        print(author_txt)
        publish=item.select_one('span.authPub.info_pub > a')
        publish_txt=publish.text
        print(publish.text)
        price=item.select_one('div.info_row.info_price > strong > em')
        price_txt=int(price.text.replace(',',''))
        print(price_txt)
        grade=item.select_one('div.info_row.info_rating > span.rating_grade > em')
        grade_txt=0.0
        if grade!=None:
            grade_txt=float(grade.text)
            print(grade_txt)
        img=item.select_one('img')
        img_src=img['data-original']
        print(img_src)

        save_image(img_src,title_txt)
        book_list.append(tuple([title_txt,author_txt,publish_txt,price_txt,grade_txt]))
```
