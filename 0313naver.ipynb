```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os
import urllib.request as req
```

```python
# 네이버 이미지 검색 페이지 URL과 검색어 설정
url = "https://www.naver.com"
query = '빈지노'

# 웹드라이버 실행 및 페이지 로드
driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)

# 검색창 찾기 (name 속성으로 찾기) 및 검색어 입력
search_box = driver.find_element(By.NAME,'query') # 검색창 찾기 (class 속성으로 찾기)
search_box.send_keys(query) # 검색창에 검색어를 입력하는 역할
# search_box.send_keys(keys.ENTER)
search_box.submit() # 검색창에 입력한 검색어로 검색하는 역할
time.sleep(5)

# 이미지 창 접속하기
image_tab = driver.find_element('link text', '이미지')
image_tab.click()
time.sleep(3)

# 이미지 스크롤을 내리는 시간 간격 설정
PAUSE_TIME=2 

#현재 페이지의 세로 스크롤의 위치
last_height = driver.execute_script("return window.scrollY")
new_height = 0
interval = 2

# 현재 페이지 전체 스크롤 가능 높이
prev_height = driver.execute_script("return document.body.scrollHeight")

# 페이지 끝까지 스크롤 내리기
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # 스크롤을 내리는 역할
    time.sleep(interval) # 스크롤 내리는 시간 간격

    curr_height = driver.execute_script("return document.body.scrollHeight") # 현재 페이지 전체 스크롤 가능 높이

    if curr_height == prev_height: # 스크롤이 더 이상 되지 않을 때 
        break

    prev_height = curr_height # 현재 페이지 전체 스크롤 가능 높이를 이전 페이지 전체 스크롤 가능 높이로 변경

# 이미지 태그 찾기
img_elements = driver.find_elements(By.CSS_SELECTOR, 'img._fe_image_tab_content_thumbnail_image') 

# 이미지 저장 경로 설정
save_path = r"C:\BIGDATA\data\naver\{query}" 
import os
if not os.path.exists(save_path):
    os.makedirs(save_path)

imgs=[]

# 이미지 다운로드 및 저장
for index, img in enumerate(img_elements): 
    try:
        img_src = img.get_attribute('src')
        img_alt = img.get_attribute('alt')  

        if img_src and "http" in img_src:
            imgs.append({'src':img_src,'alt':img_alt})
            req.urlretrieve(img_src, f"{save_path}/{query}_{index}.png")
            print(f"{index}:다운로드 완료 - {img_alt}")
    except: 
        pass

        # print(img_src)
        # print(img_alt)
        
# 웹드라이버 종료
driver.close()
```

```python
# 검색어를 리스트로 해서 여러 검색어를 for문을 활용해서 이미지 파일들 저장하기

# 검색어 리스트 설정
queries = ['빈지노', '아이유', 'BTS']

# 웹드라이버 실행
driver = webdriver.Chrome()

for query in queries:
    # 네이버 이미지 검색 페이지 URL과 검색어 설정
    url = "https://www.naver.com"
    
    # 페이지 로드
    driver.get(url)
    time.sleep(3)

    # 검색창 찾기 (name 속성으로 찾기) 및 검색어 입력
    search_box = driver.find_element(By.NAME, 'query')
    search_box.send_keys(query)
    search_box.submit()  # 검색창에 입력한 검색어로 검색 실행
    time.sleep(3)

    # 이미지 탭 클릭
    image_tab = driver.find_element(By.LINK_TEXT, '이미지')
    image_tab.click()
    time.sleep(3)

    # 이미지 스크롤을 내리는 시간 간격 설정
    PAUSE_TIME = 2

    # 현재 페이지의 세로 스크롤의 위치
    last_height = driver.execute_script("return window.scrollY")
    new_height = 0
    interval = 2

    # 현재 페이지 전체 스크롤 가능 높이
    prev_height = driver.execute_script("return document.body.scrollHeight")

    # 페이지 끝까지 스크롤 내리기
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(interval)

        curr_height = driver.execute_script("return document.body.scrollHeight")

        if curr_height == prev_height:  # 스크롤이 더 이상 되지 않을 때
            break

        prev_height = curr_height  # 현재 페이지 전체 스크롤 가능 높이를 이전 페이지 전체 스크롤 가능 높이로 변경

    # 이미지 태그 찾기
    img_elements = driver.find_elements(By.CSS_SELECTOR, 'img._fe_image_tab_content_thumbnail_image')
    print(img_elements)

    # 이미지 저장 경로 설정
    save_path = f"C:/BIGDATA/data/naver2/{query}"
    if not os.path.exists(save_path):
        os.makedirs(save_path)

    imgs = []

    # 이미지 다운로드 및 저장
    for index, img in enumerate(img_elements):
        try:
            img_src = img.get_attribute('src')
            if not img_src:
                img_src = img.get_attribute('data-src')  # src 속성이 없을 경우 data-src 속성 확인
            img_alt = img.get_attribute('alt')

            if img_src and "http" in img_src:
                imgs.append({'src': img_src, 'alt': img_alt})
                req.urlretrieve(img_src, f"{save_path}/{query}_{index}.png")
                print(f"{query} - {index}: 다운로드 완료 - {img_alt}")
        except Exception as e:
            print(f"오류 발생: {e}")

# 웹드라이버 종료
driver.close()
```
