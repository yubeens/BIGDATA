```python
import os
import sys
import urllib.request
import json
import re
import pandas as pd

client_id = "Lx0yyU01fYnjVu3WWZRy"
client_secret = "GRlWh1s22_"
search_list = ['AI', 'python', '한국사']
item_list = []

for search in search_list:
    encText = urllib.parse.quote(search)
    for page in range(1, 11):
        url = f"https://openapi.naver.com/v1/search/book?query={encText}&start={page}&display=20"  # JSON 결과
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id", client_id)
        request.add_header("X-Naver-Client-Secret", client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()  # 읽기에 성공하면 200을 리턴함
        if rescode == 200:
            response_body = response.read()
            result = json.loads(response_body.decode('utf-8'))  # response 형태로 불러온 데이터이므로 load가 아니라 loads라고 함.
            items = result['items']
            for item in items:
                title = re.sub(r'<.*?>', '', item['title'])
                link = item['link']
                image = item['image']
                author = item['author']
                discount = item['discount']
                publisher = item['publisher']
                pubdate = item['pubdate']
                isbn = item['isbn']
                description = re.sub(r'<.*?>', '', item['description'])
                item_list.append((title, link, image, author, discount, publisher, pubdate, isbn, description))
        else:
            print("Error Code:" + rescode)

# DataFrame 생성 및 CSV 파일로 저장
book_df = pd.DataFrame(item_list, columns=["title", "link", 'image', 'author', 'discount', 'publisher', 'pubdate', 'isbn', 'description'])
book_df.to_csv("books.csv", index=False)
```

```python
search_keywords = ["리바이","에렌","피크핑거","미카사"]

# 한 번의 요청에서 가져올 데이터 개수 (최대 100)
displayTimes = 100

# 결과를 저장할 디렉토리
output_dir = "data"
os.makedirs(output_dir, exist_ok=True)  # data 폴더가 없으면 생성

# 검색어별 데이터 처리
for keyword in search_keywords:
    print(f"### 검색어: {keyword} 처리 중... ###")
    encText = urllib.parse.quote(keyword)  # 검색어를 UTF-8로 인코딩

    # 전체 결과를 저장할 리스트
    all_results = []

    # 페이지네이션 처리
    for resultNum in range(1, 1001, displayTimes):  # 최대 1000건까지, 100개 단위로 검색
        url = f"https://openapi.naver.com/v1/search/book?query={encText}&display={displayTimes}&start={resultNum}"
        request = urllib.request.Request(url)

        # API 인증 헤더 추가
        request.add_header("X-Naver-Client-Id", client_id)
        request.add_header("X-Naver-Client-Secret", client_secret)

        try:
            # API 요청
            response = urllib.request.urlopen(request)
            rescode = response.getcode()

            if rescode == 200:
                response_body = response.read()
                result = json.loads(response_body.decode('utf-8'))
                items = result.get('items', [])

                # API 결과 데이터를 리스트에 추가
                all_results.extend(items)
            else:
                print(f"Error: Response code {rescode}")
                break
        except Exception as e:
            print(f"Error occurred for resultNum={resultNum}: {e}")
            break

    # 검색 결과를 데이터프레임으로 변환
    df = pd.DataFrame(all_results)

    # 필요한 컬럼만 선택하여 정리
    df = df[
        ['title', 'author', 'publisher', 'pubdate', 'isbn', 'discount', 'description', 'link', 'image']
    ]

    # CSV 파일로 저장
    output_file = os.path.join(output_dir, f"{keyword}_results.csv")
    df.to_csv(output_file, index=False, encoding='utf-8-sig')  # 한글 깨짐 방지 위해 'utf-8-sig'
    print(f"'{output_file}'에 검색 결과 저장 완료.")
```
